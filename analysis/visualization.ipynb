- matplotlib ( basic visualization)
- seaborn ( advanced visualization)
- plotly ( interactive visualization)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle= '' , marker = 's', markersize=8, alpha= 0.5,figsize= (15,7), markerfacecolor = 'red', markeredgecolor ='black')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar', figsize=(15,5), color='green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='area', figsize=(15,5), color='green', alpha= 0.5)
```

```python
penguins.plot(kind= 'scatter', x ='bill_length_mm', y ='bill_depth_mm', s = 100 , alpha = 0.5)
```

```python
penguins.plot(kind= 'scatter', x ='bill_length_mm', y ='flipper_length_mm', s = 100 , alpha = 0.5)
```

- subplots

```python
fig, axs = plt.subplots(1,2, figsize=(15,4), constrained_layout = True)
penguins.plot(kind= 'scatter', x ='bill_length_mm', y ='flipper_length_mm', s = 100 , alpha = 0.5, ax= axs[0], color = 'red')
penguins.plot(kind= 'scatter', x ='bill_length_mm', y ='bill_depth_mm', s = 100 , alpha = 0.5,  ax= axs[1])
plt.show
```

```python
penguins['species'].value_counts()
```

```python
penguins['species'].value_counts().plot(kind='pie',autopct= '%1.1f%%',figsize=(5,5), title ='species', ylabel='')
```

```python
penguins['species'].value_counts().plot(kind='pie',autopct= '%1.1f%%',figsize=(5,5), title ='species', ylabel='',wedgeprops = dict(width =0.5))
```

- advanced visualization

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(data=tips, x = 'sex', hue = 'size')
```

```python
sns.displot(data=tips, x='total_bill', binwidth = 5) # distribution plot (histogram)
```

```python
sns.displot(data= tips, x ='total_bill', col= 'smoker', row= 'sex')
```

```python
sns.displot(data= tips, x ='total_bill', col= 'smoker', row= 'sex' , aspect=1, height= 3)
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="sex", multiple="dodge")
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="ecdf")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```

```python

sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", binwidth=(2, .5), cbar=True)
```

```python
sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde"
)
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
g = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

```python
sns.relplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
sns.rugplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.pairplot(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
